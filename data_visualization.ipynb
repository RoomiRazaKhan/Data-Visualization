```python
import pandas as pd
```

```python
file_path = "F:\\Roomi Raza\\Data_visualization\\data\\undernourishment_data.csv"
```

```python
data = pd.read_csv(file_path)
```

```python
column_name = "2.1.1 Prevalence of undernourishment"
```

```python
frequency_distribution = data[column_name].value_counts().reset_index()
```

```python
frequency_distribution.columns = [column_name, 'Frequency']
```

```python
print(frequency_distribution)
```

```python
data_with_frequency = data.merge(frequency_distribution, how='left', left_on=column_name, right_on=column_name)
```

```python
output_file_path = 'F:\\Roomi Raza\\Data_visualization\\data\\undernourishment_data.csv'
```

```python
data_with_frequency.to_csv(output_file_path, index=False)
```
